```python
year = [2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020]
popAf = [60,62,78,85,89,91,89,90,94,98,104]
popT = [70,72,74,72,76,76,78,89,88,80,90]
```

```python
plt.figure(figsize=(12,8),dpi=(80))
plt.xlabel("Year")
plt.ylabel("Population of Afghanistan")
plt.plot(year,popAf,ls=':',marker='+',mew=4)
plt.plot(year,popT,ls='--',lw=1)
plt.title("Afghanistan")
plt.yticks([60,70,80,90,100,110],['60m','70m','80m','90m','100m','110m'])
# adding legend for know which line is for afghanistan and which one is for turkey.
plt.legend(['Afghan','Turkey'],loc='best')
# adding an arrow to show a point using anntate
plt.annotate("Afghanistan in 2006",xytext=(2007.5,89),xy=(2008,89),arrowprops={
    'facecolor':'silver',
    'width':0.5
},fontsize=15)
```

```python
plt.subplot(1,2,1)
plt.plot(year,popAf)
plt.title("Afghanistan population")
plt.figure(figsize=(8,5),dpi=(80))
plt.subplot(1,2,2)
plt.plot(year,popT)
plt.title('Turkey Population')
# plt.figure(figsize=(5,4),dpi=(80))
```
